```python
# accuracy plot
epochs = [1, 5, 10, 20]
plt.plot(epochs, [42.68, 55.12, 72.85, 71.89], marker='.', label='AlexNet')
plt.plot(epochs, [41.40, 54.22, 70.88, 66.88], marker='.', label='VGG-16')
plt.plot(epochs, [50.18, 56.81, 75.94, 75.94], marker='.', label='ResNet-50')
plt.plot(epochs, [46.91, 58.70, 77.74, 77.77], marker='.', label='ResNet-50')
plt.xlabel('Epochs')
plt.ylabel('Test Accuracy (%)')
plt.title('CNN Model Test Accuracy')
plt.legend(loc='lower right')
plt.show()
```

```python
cnn_fps = pd.read_csv('/content/drive/MyDrive/DeepLearning/MilitaryObjectDetection/videos/fps.csv')
rcnn_fps = np.mean([11.758746441391992, 11.823181934398336, 11.804960636253586, 11.7944707350801, 11.680888318473002, 11.680888318473002, 12.36946169920343, 12.296583966170811, 11.813129913233338,11.820950380545133])

average_fps = {}
for model in cnn_fps.columns[1:]:
    average_fps[model] = np.mean(cnn_fps[model])
average_fps['faster_rcnn_model_0.pth'] = rcnn_fps

configs = list(average_fps.keys())
speedups = list(average_fps.values())

plt.figure(figsize=(8, 6))
plt.bar(configs, speedups, color='skyblue')
plt.xlabel('Configurations')
plt.ylabel('FPS')
plt.title('FPS Comparison Between Configurations for Video Processing Task')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
